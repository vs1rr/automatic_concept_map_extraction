```python
import os
import json
import numpy as np
import regex as re
import matplotlib.pyplot as plt

class MetricsPlotter:
    def __init__(self, root_dir):
        self.root_dir = root_dir
        self.metrics_data = [] 
        self.top_10_precision = {}
        self.top_10_recall = {}
        self.top_10_f1 = {}

    def read_json(self, file_path):
        with open(file_path, 'r') as file:
            return json.load(file)
            
    def sanitize_filename(self, filename):
        filename = filename.replace(' ', '_')
        filename = re.sub(r'[^\w\-_\.]', '', filename)
        return filename
        
    def get_params_string(self, params):
        summary = params.get('summary', {})
        ranking = params.get('ranking', {})
        entity = params.get('entity', {})
        relation = params.get('relation', {})

        preprocess_str = f"Preprocess: {params.get('preprocess', None)}" if params.get('preprocess', None) is not None else ''
        summary_method = f"Summary Method: {summary.get('summary_method', None)}" if summary.get('summary_method', None) is not None else ''
        summary_percentage = f"Summary Percentage: {summary.get('summary_percentage', None)}" if summary.get('summary_percentage', None) is not None else ''
        ranking_method = f"Ranking Method: {ranking.get('ranking', None)}" if ranking.get('ranking', None) is not None else ''
        ranking_perc_threshold = f"Ranking Perc Threshold: {ranking.get('ranking_perc_threshold', None)}" if ranking.get('ranking_perc_threshold', None) is not None else ''
        options_ent = f"Options Ent: {', '.join(sorted(entity.get('options_ent', [None])))}" if sorted(entity.get('options_ent', [None])) != [None] else ''
        options_rel = f"Options Rel: {', '.join(sorted(relation.get('options_rel', [None])))}" if sorted(relation.get('options_rel', [None])) != [None] else ''
        local_rm = f"Local RM: {relation.get('local_rm', None)}" if relation.get('local_rm', None) is not None else ''
        rebel_model = f"Rebel Model: {relation.get('rebel_model', None)}" if relation.get('rebel_model', None) is not None else ''
        
        params_list = [preprocess_str, summary_method, summary_percentage, ranking_method, ranking_perc_threshold, options_ent, options_rel, local_rm, rebel_model]
        params_str = '\n'.join(filter(None, params_list))

        return params_str

    def compute_average_meteor_scores(self, json_data):
        meteor_precision_scores = []
        meteor_recall_scores = []
        meteor_f1_scores = []
        for v in json_data.values():
            if 'meteor' in v:
                meteor_scores = v['meteor']
                if 'precision' in meteor_scores and 'recall' in meteor_scores and 'f1' in meteor_scores:
                    meteor_precision_scores.append(meteor_scores['precision'])
                    meteor_recall_scores.append(meteor_scores['recall'])
                    meteor_f1_scores.append(meteor_scores['f1'])
                else:
                    print("Warning: Missing Meteor scores in JSON data.")
            else:
                print("Warning: Missing Meteor scores in JSON data.")
        avg_precision = np.mean(meteor_precision_scores)
        avg_recall = np.mean(meteor_recall_scores)
        avg_f1 = np.mean(meteor_f1_scores)
        return avg_precision, avg_recall, avg_f1
        
    def print_metrics(self, json_file, params_str):
        with open(json_file) as f:
            data = json.load(f)
        avg_precision, avg_recall, avg_f1 = self.compute_average_meteor_scores(data)
        self.metrics_data.append({
            "Params": params_str,
            "Avg Meteor Precision": avg_precision,
            "Avg Meteor Recall": avg_recall,
            "Avg Meteor F1": avg_f1
        })
        # Debugging statement to check the structure
        print(f"Appended metrics: {self.metrics_data[-1]}")

    def plot_metrics_bar_directory(self):
        for root, dirs, _ in os.walk(self.root_dir):
            for dir_name in dirs:
                metrics_path = os.path.join(root, dir_name, 'metrics.json')
                params_path = os.path.join(root, dir_name, 'params.json')
    
                if all(os.path.exists(p) for p in [metrics_path, params_path]):
                    try:
                        params = self.read_json(params_path)
                        params_str = self.get_params_string(params)
                        self.print_metrics(metrics_path, params_str)  # Uncommented this line
                    except Exception as e:
                        print(f"Error processing files in directory: {os.path.join(root, dir_name)}")
                        print(e) 

    def print_metrics_data(self):
            for params, metrics in self.metrics_data.items():
                print(f"Params: {params}")
                print(f"Metrics: {metrics}")

    # def sort_and_store_top_metrics(self):
    #     sorted_by_precision = sorted(self.metrics_data, key=lambda x: x["Avg Meteor Precision"], reverse=True)[:10]
    #     print(f"Top 10 by Precision: {sorted_by_precision}")
    #     sorted_by_recall = sorted(self.metrics_data, key=lambda x: x["Avg Meteor Recall"], reverse=True)[:10]
    #     print(f"Top 10 by Recall: {sorted_by_recall}")
    #     sorted_by_f1 = sorted(self.metrics_data, key=lambda x: x["Avg Meteor F1"], reverse=True)[:10]
    #     print(f"Top 10 by F1: {sorted_by_f1}")

    #     self.top_10_precision = {item["Params"]: item["Avg Meteor Precision"] for item in sorted_by_precision}
    #     self.top_10_recall = {item["Params"]: item["Avg Meteor Recall"] for item in sorted_by_recall}
    #     self.top_10_f1 = {item["Params"]: item["Avg Meteor F1"] for item in sorted_by_f1}

    def sort_and_store_top_metrics(self):
        sorted_by_precision = sorted(self.metrics_data, key=lambda x: x["Avg Meteor Precision"])
        sorted_by_recall = sorted(self.metrics_data, key=lambda x: x["Avg Meteor Recall"])
        sorted_by_f1 = sorted(self.metrics_data, key=lambda x: x["Avg Meteor F1"])
    
        self.best_worst_precision = {
            sorted_by_precision[0]["Params"]: sorted_by_precision[0]["Avg Meteor Precision"],
            sorted_by_precision[-1]["Params"]: sorted_by_precision[-1]["Avg Meteor Precision"]
        }
        self.best_worst_recall = {
            sorted_by_recall[0]["Params"]: sorted_by_recall[0]["Avg Meteor Recall"],
            sorted_by_recall[-1]["Params"]: sorted_by_recall[-1]["Avg Meteor Recall"]
        }
        self.best_worst_f1 = {
            sorted_by_f1[0]["Params"]: sorted_by_f1[0]["Avg Meteor F1"],
            sorted_by_f1[-1]["Params"]: sorted_by_f1[-1]["Avg Meteor F1"]
        }
        
    def print_top_metrics(self):
        print("Top 10 Avg Meteor Precision:")
        for params, value in self.top_10_precision.items():
            print(f"Params: {params}\nPrecision: {value}\n")
        
        print("Top 10 Avg Meteor Recall:")
        for params, value in self.top_10_recall.items():
            print(f"Params: {params}\nRecall: {value}\n")
        
        print("Top 10 Avg Meteor F1:")
        for params, value in self.top_10_f1.items():
            print(f"Params: {params}\nF1: {value}\n")

    def plot_top_metrics(self):
        fig_dir = 'plot/plot_metrics/meteor/best_worst_meteor'
        if not os.path.exists(fig_dir):
            os.makedirs(fig_dir)
            
        sns.set(style="whitegrid")
    
        def plot_metric(metric_dict, title, ylabel):
            params = list(metric_dict.keys())
            values = list(metric_dict.values())
    
            plt.figure(figsize=(12, 6))
            barplot = sns.barplot(x=params, y=values, palette="viridis")
            barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45, ha='right', fontsize=10)
            barplot.set_title(title, fontsize=16)
            barplot.set_xlabel('Parameters', fontsize=12)
            barplot.set_ylabel(ylabel, fontsize=12)
            for index, value in enumerate(values):
                barplot.text(index, value, f'{value:.2f}', color='black', ha="center", fontsize=10)
            plt.tight_layout()
            fig_path = os.path.join(fig_dir, f"{title.replace(' ', '_')}_plot.png")
            plt.savefig(fig_path)
            plt.show()
            plt.close()
    
        plot_metric(self.best_worst_f1, "Best_and_Worst_Avg_Meteor_F1_Scores", "F1 Score")
        plot_metric(self.best_worst_precision, "Best_and_Worst_Avg_Meteor_Precision_Scores", "Precision Score")
        plot_metric(self.best_worst_recall, "Best_and_Worst_Avg_Meteor_Recall_Scores", "Recall Score")
```
